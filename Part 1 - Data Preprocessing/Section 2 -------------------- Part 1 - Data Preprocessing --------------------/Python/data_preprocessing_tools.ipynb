# Data Preprocessing Tools

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler, Normalizer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
x = dataset.drop(columns = 'Purchased').values
y = dataset.iloc[:, -1].to_numpy()
```

## Taking care of missing data

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
x[:, 1:] = imputer.fit_transform(x[:, 1:])
```

## Encoding categorical data

### Encoding the Independent Variable

```python
ct = ColumnTransformer(transformers=[('label', OneHotEncoder(), [0])], remainder='passthrough')
x = ct.fit_transform(x)
```

```python
print(x)
```

### Encoding the Dependent Variable

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

## Feature Scaling

```python
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```
